```python
#q1
from matplotlib import pyplot as plt
import numpy as np


fig1 =  plt.figure()
ax = fig1.add_axes([0,0,1,1])
x = [0,100]
y = [0,200]
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('q1')
```

```python
#q2
fig2 = plt.figure()

ax1 = fig2.add_axes([0,0,1,1])

ax2 = fig2.add_axes([0.2,0.5,0.2,0.2])

x= [0,2,4,6,8,10]
y= [0,4,16,36,64,100]

ax1.plot(x,y)

ax2.plot(y,x)

ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.set_xlabel('y')
ax2.set_ylabel('x')

ax1.set_title('q2')
```

```python
#q3
import csv
import pandas as pd

company = pd.read_csv("/home/student/Desktop/company_sales_data.csv")
company_df = pd.DataFrame(company)
print(company_df)
month =  company_df['month_number']
profit = company_df['total_profit']

fig3 = plt.figure()

ax3 = fig3.add_axes([0,0,1,1])
ax3.set_xlabel('Month Number')
ax3.set_ylabel('Total Profit')
ax3.set_title('q3')

ax3.plot(month,profit)
```

```python
#q4

plt.plot(month, profit , linestyle= '--' , color='red', linewidth = 3 , marker = 'o', mfc='k' ,label ="Profit data of last year",)
plt.xlabel("Month Number")
plt.ylabel("Sold units nuber")
plt.title("Company Sales Data of Last Year")
plt.legend( loc = 4)
plt.ylim(100000)
plt.xlim()
plt.show()
```
